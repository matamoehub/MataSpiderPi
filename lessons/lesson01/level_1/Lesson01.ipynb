This level is just about learning how SpiderPi moves.

## Movement Goals

Practice these body commands:
- forward
- backward
- left
- right
- turn_left
- turn_right
- stop

```python
from lesson_header import *

bot.help()
```

```python
bot.body.forward(0.6)
bot.body.stop()
```

```python
bot.body.backward(0.6)
bot.body.stop()
```

```python
bot.body.left(0.6)
bot.body.right(0.6)
bot.body.stop()
```

```python
bot.body.turn_left(0.5)
bot.body.turn_right(0.5)
bot.body.stop()
```

```python
bot.display.text("I can move", "around now", seconds=1.0)
bot.body.wave()
```
