# SpiderPi Show Time

This demo is meant to feel like a mini robot performance, not just a hardware checklist.

## Performance Plan

SpiderPi will:
1. Make a grand entrance
2. Flash lights and draw on the displays
3. Greet the audience with Piper speech
4. Show off movement and action groups
5. Wave and pose with the arm
6. Look around and use vision
7. React to distance sensing
8. Finish with a playful finale


```python
from lesson_loader import setup
setup()
from student_robot_v2 import bot

bot.help()
```

```python
# Grand entrance
bot.display.text("SpiderPi Show", "Welcome!", "Get ready", "for robot fun")
bot.display.smile(seconds=1.5)
bot.lights.purple()
bot.sound.beep(2400, 0.12)
bot.sound.beep(2800, 0.12)
bot.buzzer.off()
bot.speech.say("Hello everyone. I am Spider Pi, and I am ready to perform.")
```

```python
# Body movement with a bit of swagger
bot.body.forward(0.6)
bot.body.turn_left(0.35)
bot.body.right(0.45)
bot.body.left(0.45)
bot.body.turn_right(0.35)
bot.body.wave()
bot.body.stop()
```

```python
# Lights, display, and camera personality moment
bot.lights.blue()
bot.display.text("Mood Change", "Blue lights", "Big smile", "Camera moves")
bot.camera.center_all()
time.sleep(0.5)
bot.camera.look_up()
time.sleep(0.5)
bot.camera.look_down()
time.sleep(0.5)
bot.camera.glance_left()
time.sleep(0.5)
bot.camera.glance_right()
time.sleep(0.5)
bot.camera.shake()
bot.display.diamond(seconds=1.5)
bot.speech.say("I can glow, smile, and move my camera too.")
```

```python
# Arm pose, turning, and gripper routine
bot.display.text("Arm Demo", "Turn left", "Turn right", "Grab")
bot.arm.home()
time.sleep(0.5)
bot.arm.turn_left()
time.sleep(0.8)
bot.arm.center_turn()
time.sleep(0.8)
bot.arm.turn_right()
time.sleep(0.8)
bot.arm.center_turn()
time.sleep(0.8)
bot.arm.open()
bot.speech.say("Watch my robot arm.")
bot.arm.move(0, 15, 12, seconds=0.5)
time.sleep(0.6)
bot.arm.close()
time.sleep(0.6)
bot.arm.carry()
time.sleep(0.6)
bot.arm.open()
time.sleep(0.6)
bot.arm.home()
```

```python
# Camera and color vision check
bot.display.text("Looking Around", "Center", "Colors", "Snapshot")
bot.camera.center_all()
time.sleep(0.5)
bot.vision.snapshot(show=True)
red_seen = bot.vision.find_color("red", show=True)
print(red_seen)
```

```python
# Face detection cameo
bot.display.text("Face Time", "MediaPipe", "Look this way", "Say hello")
face_result = bot.vision.show_faces(show=True)
print(face_result)
bot.speech.say("I can spot a face with Media Pipe.")
```

```python
# Hand gesture recognition cameo
bot.display.text("Hand Time", "Show me", "rock paper", "scissors")
hand_result = bot.vision.show_hands(show=True)
print(hand_result)
bot.speech.say("I can watch your hand gestures too.")
```

```python
# Distance reaction bit
distance_cm = bot.distance.cm()
print("Distance in cm:", distance_cm)
if bot.distance.is_close(25):
    bot.lights.yellow()
    bot.display.text("Whoa!", "Something is", "close to me", f"{distance_cm} cm")
    bot.speech.say("That is a little too close for my comfort.")
else:
    bot.lights.green()
    bot.display.text("All Clear", "Plenty of room", "for dancing", f"{distance_cm} cm")
    bot.speech.say("Plenty of space. Time for the finale.")
```

```python
# Final playful performance
bot.display.text("Finale", "Dance mode", "Lights on", "SpiderPi!")
bot.display.triangle(seconds=1.5)
bot.lights.red()
bot.sound.beep(1800, 0.08)
bot.sound.beep(2200, 0.08)
bot.sound.beep(2600, 0.12)
bot.buzzer.off()
bot.body.dance()
bot.lights.blue()
bot.body.twist()
bot.lights.green()
bot.body.kick()
bot.camera.center_all()
time.sleep(0.5)
bot.arm.home()
bot.display.smile(seconds=1.5)
bot.speech.say("Thank you for watching the Spider Pi show.")
bot.lights.off()
bot.body.stop()
```
