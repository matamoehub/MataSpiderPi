# SpiderPi Show Time

This demo is meant to feel like a mini robot performance, not just a hardware checklist.

## Performance Plan

SpiderPi will:
1. Make a grand entrance
2. Flash lights and draw on the displays
3. Greet the audience with Piper speech
4. Show off movement and action groups
5. Wave and pose with the arm
6. Look around and use vision
7. React to distance sensing
8. Finish with a playful finale


```python
from lesson_loader import setup
setup()
from student_robot_v2 import bot

bot.help()
```

```python
# Grand entrance
bot.display.text("SpiderPi Show", "Welcome!", "Get ready", "for robot fun")
bot.display.smile()
bot.lights.purple()
bot.sound.beep(2400, 0.12)
bot.sound.beep(2800, 0.12)
bot.speech.say("Hello everyone. I am Spider Pi, and I am ready to perform.")
```

```python
# Body movement with a bit of swagger
bot.body.forward(0.6)
bot.body.turn_left(0.35)
bot.body.right(0.45)
bot.body.left(0.45)
bot.body.turn_right(0.35)
bot.body.wave()
bot.body.stop()
```

```python
# Lights and display personality moment
bot.lights.blue()
bot.display.text("Mood Change", "Blue lights", "Big smile", "Tiny wiggle")
bot.camera.tiny_wiggle(seconds=1.5)
bot.display.diamond()
bot.speech.say("I can glow, smile, and wiggle too.")
```

```python
# Arm pose and gripper routine
bot.arm.home()
bot.arm.open()
bot.speech.say("Watch my robot arm.")
bot.arm.move(0, 15, 12, seconds=1.0)
bot.arm.close()
bot.arm.carry()
bot.arm.open()
bot.arm.home()
```

```python
# Camera and vision check
bot.display.text("Looking Around", "Camera centre", "Find a color", "Take a snapshot")
bot.camera.center_all()
bot.camera.glance_left()
bot.camera.glance_right()
bot.vision.snapshot(show=True)
red_seen = bot.vision.find_color("red", show=True)
print(red_seen)
```

```python
# MediaPipe face-detection cameo
bot.display.text("Face Time", "MediaPipe demo", "Look this way", "Say hello")
face_result = bot.vision.detect_faces(show=True)
print(face_result)
bot.speech.say("I can spot a face with the Spider Pi Media Pipe demo.")
```

### MediaPipe Note

MataSpiderPi now matches the MataTurboPi V2 naming style for MediaPipe calls: `bot.vision.detect_faces()`, `bot.vision.recognize_hands()`, and `bot.vision.detect_pose()`. Face detection is confirmed in the upstream SpiderPi repo, and MataSpiderPi now adds a student-friendly hand-recognition helper for gesture lessons like rock, paper, scissors. Pose detection is still not wired up in SpiderPi yet.


```python
# Distance reaction bit
distance_cm = bot.distance.cm()
print("Distance in cm:", distance_cm)
if bot.distance.is_close(25):
    bot.lights.yellow()
    bot.display.text("Whoa!", "Something is", "close to me", f"{distance_cm} cm")
    bot.speech.say("That is a little too close for my comfort.")
else:
    bot.lights.green()
    bot.display.text("All Clear", "Plenty of room", "for dancing", f"{distance_cm} cm")
    bot.speech.say("Plenty of space. Time for the finale.")
```

```python
# Final playful performance
bot.display.text("Finale", "Dance mode", "Lights on", "SpiderPi!")
bot.display.triangle()
bot.lights.red()
bot.sound.melody("C4:0.5 E4:0.5 G4:0.5 C5:1", bpm=180)
bot.body.dance()
bot.lights.blue()
bot.body.twist()
bot.lights.green()
bot.body.kick()
bot.arm.home()
bot.display.smile()
bot.speech.say("Thank you for watching the Spider Pi show.")
bot.lights.off()
bot.body.stop()
```
