# SpiderPi Show Time

This demo is meant to feel like a mini robot performance, not just a hardware checklist.

## Performance Plan

SpiderPi will:
1. Make a grand entrance
2. Walk, turn, and show body movement
3. Animate spider eyes on the LED matrix
4. Show camera and arm personality moves
5. Use color, face, and hand vision
6. React to distance sensing
7. Finish with a bigger visual finale


```python
from lesson_loader import setup
setup()
from student_robot_v2 import bot

bot.help()
```

```python
# Grand entrance
bot.display.text("SpiderPi Show", "Eyes on", "Arms ready", "Lets go", seconds=1.5)
bot.display.eyes(seconds=1.0)
bot.display.blink(seconds=0.8)
bot.display.look_left(seconds=0.8)
bot.display.look_right(seconds=0.8)
bot.lights.purple()
bot.sound.beep(2400, 0.12)
bot.sound.beep(2800, 0.12)
bot.buzzer.off()
bot.display.smile(seconds=1.2)
```

```python
# Body movement with a bit of swagger
bot.body.forward(0.6)
bot.body.turn_left(0.35)
bot.body.right(0.45)
bot.body.left(0.45)
bot.body.turn_right(0.35)
bot.body.wave()
bot.body.stop()
```

```python
# LED eyes and camera personality moment
bot.lights.blue()
bot.display.text("Spider Eyes", "Blink wink", "Look around", "Stay sharp", seconds=1.2)
bot.display.eyes(seconds=0.8)
bot.display.look_up(seconds=0.8)
bot.display.look_down(seconds=0.8)
bot.display.look_left(seconds=0.8)
bot.display.look_right(seconds=0.8)
bot.display.wink(seconds=0.8)
bot.display.sleep(seconds=0.8)
bot.display.eyes(seconds=0.8)
bot.camera.center_all()
time.sleep(0.4)
bot.camera.look_up()
time.sleep(0.4)
bot.camera.look_down()
time.sleep(0.4)
bot.camera.glance_left()
time.sleep(0.4)
bot.camera.glance_right()
time.sleep(0.4)
bot.camera.wiggle()
bot.display.diamond(seconds=1.2)
```

```python
# Arm pose, turning, and gripper routine
bot.display.text("Arm Demo", "Turn pose", "Grab move", "Reset", seconds=1.2)
bot.arm.home()
time.sleep(0.4)
bot.arm.turn_left()
time.sleep(0.5)
bot.arm.center_turn()
time.sleep(0.5)
bot.arm.turn_right()
time.sleep(0.5)
bot.arm.center_turn()
time.sleep(0.5)
bot.arm.open()
time.sleep(0.4)
bot.arm.move(0, 15, 12, seconds=0.35)
time.sleep(0.5)
bot.arm.close()
time.sleep(0.5)
bot.arm.carry()
time.sleep(0.5)
bot.arm.open()
time.sleep(0.5)
bot.arm.home()
```

```python
# Camera and color vision check
bot.display.text("Looking Around", "Center", "Colors", "Snapshot", seconds=1.2)
bot.camera.center_all()
time.sleep(0.5)
snapshot_result = bot.vision.snapshot(show=True)
print(snapshot_result)
red_seen = bot.vision.find_color("red", show=True)
print(red_seen)
```

```python
# Face detection cameo
bot.display.text("Face Time", "Find faces", "Outline", "Count", seconds=1.2)
face_result = bot.vision.show_faces(show=True)
print(face_result)
```

```python
# Hand gesture recognition cameo
bot.display.text("Hand Time", "Show me", "rock paper", "scissors", seconds=1.2)
hand_result = bot.vision.show_hands(show=True)
print(hand_result)
```

```python
# Distance reaction bit
distance_cm = bot.distance.cm()
print("Distance in cm:", distance_cm)
if bot.distance.is_close(25):
    bot.lights.yellow()
    bot.display.text("Whoa!", "Too close", "Back it up", f"{distance_cm} cm", seconds=1.4)
    bot.display.shut_eyes(seconds=1.0)
else:
    bot.lights.green()
    bot.display.text("All Clear", "Plenty of room", "For dancing", f"{distance_cm} cm", seconds=1.4)
    bot.display.eyes(seconds=1.0)
```

```python
# Final playful performance
bot.display.text("Finale", "Dance mode", "Lights on", "SpiderPi!", seconds=1.2)
bot.display.triangle(seconds=1.0)
bot.lights.red()
bot.sound.beep(1800, 0.08)
bot.sound.beep(2200, 0.08)
bot.sound.beep(2600, 0.12)
bot.buzzer.off()
bot.body.dance()
bot.lights.blue()
bot.body.twist()
bot.lights.green()
bot.body.kick()
bot.camera.center_all()
time.sleep(0.4)
bot.camera.wiggle()
time.sleep(0.4)
bot.arm.home()
bot.display.blink(seconds=0.8)
bot.display.eyes(seconds=0.8)
bot.display.smile(seconds=1.2)
bot.lights.off()
bot.body.stop()
```
